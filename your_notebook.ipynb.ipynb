```python
# Install (safe even if already installed)
!pip install pandas numpy faker

import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Initialize
fake = Faker()
num_records = 200
start_time = datetime.now()

data = []

for i in range(num_records):
    patient_id = f"P{random.randint(1000, 9999)}"
    timestamp = start_time + timedelta(minutes=i)

    heart_rate = random.randint(60, 100)
    systolic_bp = random.randint(110, 140)
    diastolic_bp = random.randint(70, 90)
    temperature = round(random.uniform(36.0, 37.5), 1)
    oxygen_level = random.randint(95, 100)
    steps = random.randint(0, 50)

    data.append({
        "Patient_ID": patient_id,
        "Timestamp": timestamp,
        "Heart_Rate": heart_rate,
        "Blood_Pressure": f"{systolic_bp}/{diastolic_bp}",
        "Temperature_C": temperature,
        "Oxygen_Level_%": oxygen_level,
        "Steps": steps
    })

# Create DataFrame
df = pd.DataFrame(data)

# Show preview
print("Preview:")
print(df.head())

# Check structure
print("\nInfo:")
print(df.info())

# Save file
df.to_csv("smart_healthcare_data.csv", index=False)

# Download file
from google.colab import files
files.download("smart_healthcare_data.csv")
```
